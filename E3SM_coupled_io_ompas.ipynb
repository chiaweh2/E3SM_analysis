```python
histlist = ['3d_monthly','forcing_daily','diagnostic_daily']

varlist_1 = ['salinity', 'temperature', 'velocityZonal', 'velocityMeridional']

varlist_2 = [ "windStressZonal", "windStressMeridional", "latentHeatFlux",
             "sensibleHeatFlux", "longWaveHeatFluxUp", "longWaveHeatFluxDown",
             "shortWaveHeatFlux", "evaporationFlux", "rainFlux", "riverRunoffFlux"]

varlist_3 = ["tThreshMLD", "dThreshMLD", "dGradMLD", "tGradMLD", "ssh", 
             "pressureAdjustedSSH", "boundaryLayerDepth", "salinityAtSurface",
             "temperatureAtSurface", "surfaceVelocityMeridional","surfaceVelocityZonal"]


varlist = [varlist_1,varlist_2,varlist_3]
```

```python
import importlib
import io_model
importlib.reload(io_model)
from io_model import E3SM_daily_cori_io_dask

output_dir = '/maloney-scratch/joedhsu/proj1/data/E3SM_simulation/20211029_mod_coare30.HIST2000_branched_all.A_WCYCL20TRS_CMIP6.ne30_oECv3_ICG.cori-knl/'

for nhist,hist in enumerate(histlist):
    print("Collecting %s data"%hist)
    ds_e3sm_h = E3SM_daily_cori_io_dask(hist=hist,case='10years',realm='ocn')
    for var in varlist[nhist]:
        print('output %s'%var)
        print('Located in %s'%output_dir)
        ds_e3sm_h[var].to_netcdf(output_dir+'HIST2000_branched_coare30_mpaso_%s.nc'%var)
    
        
```
